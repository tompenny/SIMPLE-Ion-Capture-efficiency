```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xl = np.linspace(0.001, 0.04, 100)
yl = np.linspace(0.001, 0.02, 100)
zl = np.linspace(0.001, 0.02, 100)
SAl = []
d = 0.005
t = 0.006604
for x in xl:
    for y in yl:
        for z in zl:
            h = np.sqrt(x**2 + y**2 + z**2)
            h2 = np.sqrt((x+t)**2 + (np.sqrt(y**2 + z**2)+d/2)**2)
            if (np.sqrt(z**2+y**2)+d/2)**2/h2**2 > d**2/(d**2+t**2):
                SAl.append(0)
            else:
                if z < d/2:
                    theta = np.arctan((z+d/2)/np.sqrt((x+t)**2+y**2)) - np.arctan((z-d/2)/np.sqrt((x+t)**2+y**2))
                else:
                    theta = np.arctan((z+d/2)/np.sqrt((x+t)**2+y**2)) - np.arctan((z-d/2)/np.sqrt((x)**2+y**2))
                if y < d/2:
                    phi = np.arctan((y+d/2)/np.sqrt((x+t)**2+z**2)) - np.arctan((y-d/2)/np.sqrt((x+t)**2+z**2))
                else:
                    phi = np.arctan((y+d/2)/np.sqrt((x+t)**2+z**2)) - np.arctan((y-d/2)/np.sqrt((x)**2+z**2))
                a = h*np.tan(theta)
                b = h*np.tan(phi)
                zeta = b/a # this is tan(theta)/tan(phi)
                beta = b/h # this is actually tan(theta)
                if zeta > 1:
                    zeta = a/b
                    beta = a/h
                k = np.sqrt(beta**2*(1-zeta**2)/(beta**2+zeta**2))
                theta2 = np.linspace(0, np.pi/2, 1000)
                func = ((1-(1-zeta**2)*np.sin(theta2)**2)*np.sqrt(1-k**2*np.sin(theta2)**2))**(-1)
                integ = np.trapz(func, theta2)
                I = 4*zeta**2/np.sqrt(zeta**2 + beta**2)*integ 
                SA = 2*np.pi - I
                SAl.append(SA)
```

```python
4*np.pi/np.mean(SAl)
```

```python
fig, ax = plt.subplots(5, 2, figsize = (12, 25))
for i in range(0, 10, 2):
    for j in range(0, 2, 1):
        im = ax[int(i/2), j].imshow(np.reshape(SAl[10000*(i+j)*10:((i+j)*10+1)*10000], (100,100))/4/np.pi, extent = (0, 0.02, 0.02, 0))
        ax[int(i/2), j].set_xlabel('z (m)')
        ax[int(i/2), j].set_ylabel('y (m)')
        ax[int(i/2), j].set_title('x = ' + str(xl[10*(i+j)]) + '(m)')
        plt.colorbar(im, ax = ax[int(i/2), j], label = 'Solid Angle (steradian)')
    #plt.colorbar()
plt.show()
```
